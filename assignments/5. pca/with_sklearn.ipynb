Implement PCA using sklearn library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('diabetes.csv')
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca_sklearn = pca.fit_transform(data_scaled)

df_pca_sklearn = pd.DataFrame(X_pca_sklearn, columns=['PC1', 'PC2'])
df_pca_sklearn["Outcome"] = data["Outcome"]
print(df_pca_sklearn)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df_pca_sklearn["PC1"], y=df_pca_sklearn["PC2"], hue=df_pca_sklearn["Outcome"])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection (Sklearn)")
plt.show()
```
